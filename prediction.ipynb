```python
for i in ['Season','Month','Deasaster Happen in last 3month','Vegetable condition']:
    dummies=pd.get_dummies(df[f'{i}'])
    df=pd.concat([df,dummies],axis = 1)
    del df[f'{i}']
```

```python
del df['Vegetable']
```

```python
data=df.columns
len(data)
X=df[[data[x] for x in range(len(data)) if x != 1]].values
pd.DataFrame(X)
y=df['Price per kg'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=1)
y_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
r2=metrics.r2_score(y_test,y_pred)
mae=metrics.mean_absolute_error(y_test,y_pred)
mse=metrics.mean_squared_error(y_test,y_pred,squared=True)
rmse=metrics.mean_squared_error(y_test,y_pred,squared=False)
print(r2,mse)
```
